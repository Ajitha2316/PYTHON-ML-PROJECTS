```python
# Load the wine dataset
data = load_wine()

# Create a DataFrame with data and feature names
df = pd.DataFrame(data.data, columns=data.feature_names)

# Optionally, add the target column if you want it in the CSV
df['target'] = data.target

# Save to CSV
df.to_csv("wine_data.csv", index=False)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X
```

```python
from sklearn.decomposition import PCA
```

```python
PCA_model=PCA(n_components=2)
```

```python
PCA_model.fit(scaled_X)
```

```python
PCA_model.transform(scaled_X)
```

```python
#get results of fiting and transformation#
pc_results=PCA_model.fit_transform(scaled_X)
len(pc_results)
```

```python
explained_variance = []
for n in range(1, 13):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,13), explained_variance)
plt.ylabel('Explained Variance')
plt.xlabel('No of components')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create the 3D plot
plt.figure(figsize=(8, 6), dpi=150)
ax = plt.axes(projection='3d')

# Plot the PCA data with color by target and a colormap
scatter = ax.scatter3D(pca_pixels[:, 0], pca_pixels[:, 1], pca_pixels[:, 2], 
                       c=wine_dictionary['target'], cmap='viridis')

# Add labels
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')

# Add color bar
plt.colorbar(scatter, ax=ax, label="Wine Target")
plt.show()
```
