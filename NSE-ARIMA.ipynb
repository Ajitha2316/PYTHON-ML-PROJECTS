```python
data = pd.read_csv("C:/Users/AJITHA REDDY/Downloads/NSE/DATA FOR ANALYSIS/NIFTY 50.csv")

# Step 2: Preprocessing
# Ensure the dataset has a 'Date' and 'Close' column
if 'Date' not in data.columns or 'Close' not in data.columns:
    raise ValueError("Dataset must contain 'Date' and 'Close' columns.")

# Convert 'Date' to datetime and sort by date
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data = data.dropna(subset=['Date']).sort_values('Date')

# Drop rows with missing 'Close' values
data = data.dropna(subset=['Close'])

# Set 'Date' as the index
data.set_index('Date', inplace=True)

# Step 3: Split data into train and test sets
train_size = int(len(data) * 0.8)
train, test = data.iloc[:train_size], data.iloc[train_size:]

# Step 4: Automatically determine ARIMA order using auto_arima
arima_model = auto_arima(train['Close'], seasonal=False, trace=True, suppress_warnings=True)
best_order = arima_model.order
print(f"Best ARIMA order: {best_order}")

# Step 5: Fit the ARIMA model with the determined order
model = ARIMA(train['Close'], order=best_order)
arima_result = model.fit()

# Step 6: Make predictions
forecast = arima_result.forecast(steps=len(test))
test['Predicted'] = forecast.values

# Calculate Metrics
mae = mean_absolute_error(test['Close'], test['Predicted'])
mse = mean_squared_error(test['Close'], test['Predicted'])
rmse = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

# Step 7: Plot actual vs predicted values
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['Close'], label='Train Data', color='blue')
plt.plot(test.index, test['Close'], label='Test Data', color='green')
plt.plot(test.index, test['Predicted'], label='Predicted Data', color='orange')
plt.title('ARIMA Model - Actual vs Predicted')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()
plt.show()

# Step 8: Forecast for the next 2 years
future_steps = 365 * 2  # Forecast for 2 years
future_forecast = arima_result.get_forecast(steps=future_steps)
future_forecast_df = future_forecast.conf_int()
future_forecast_df['Forecasted_Close'] = future_forecast.predicted_mean
future_forecast_df.index = pd.date_range(start=data.index[-1] + timedelta(days=1), periods=future_steps)

# Step 8: Forecast for the next 720 days
future_steps = 720  # Forecast for 720 days
future_forecast = arima_result.get_forecast(steps=future_steps)

# Extract forecasted mean and confidence intervals
future_forecast_df = future_forecast.conf_int()
future_forecast_df['Forecasted_Close'] = future_forecast.predicted_mean
future_forecast_df.index = pd.date_range(start=data.index[-1] + timedelta(days=1), periods=future_steps)

# Create a DataFrame with date and predicted close prices
future_prices_df = pd.DataFrame({
    'Date': future_forecast_df.index,
    'Predicted_Close': future_forecast_df['Forecasted_Close']
})

# Display the first 30 rows of the 720-day forecast
print(future_prices_df.head(30))

# Display the first 10 rows of the confidence intervals and predicted prices
print(confidence_interval_df.head(10))

# Optionally save the confidence interval data to a CSV file
confidence_interval_df.to_csv("Future_Predicted_Confidence_Intervals.csv", index=False)


# Step 9: Plot future forecast with confidence intervals
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='Historical Data', color='blue')
plt.plot(future_forecast_df.index, future_forecast_df['Forecasted_Close'], label='Future Forecast', color='orange')
plt.fill_between(
    future_forecast_df.index,
    future_forecast_df.iloc[:, 0],
    future_forecast_df.iloc[:, 1],
    color='orange',
    alpha=0.2,
    label='Confidence Interval'
)
plt.title('NIFTY 50 ARIMA Model')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()
plt.show()
```
